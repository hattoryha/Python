## 3/

```python
x = np.linspace(-2 * np.pi, 2 * np.pi)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.tan(x))
plt.legend(['sin(x)', 'cos(x)', 'tan(x)'])
plt.title('Graphs of sin(x) and cos(x), tan(x) within -2pi -> 2pi')
```

##  gererate 1000 random integers corresponding to 1000 days from 09/02/2010

```python
days = pd.date_range('2010/02/09', periods = 1000);

data=np.random.randint(-100,100, 1000)

s = pd.Series(data,index = days)

data_frame = np.cumsum(s)


data_frame.plot()
```
